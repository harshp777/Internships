```python
#BIS
mi_bats=['Ishan Kishan' ,'KH Pandya', 'HH Pandya', 'Q de Kock', 'KA Pollard', 'RG Sharma']



for i in mi_bats:
    


    tot_runs=d15[(d15['batting_team']=='Mumbai Indians') & (d15['batsman']==i)]['batsman_runs'].sum()


    bound_runs=d15[(d15['batting_team']=='Mumbai Indians') & (d15['batsman']==i) &((d15['batsman_runs']==4) | (d15['batsman_runs']==6))]['batsman_runs'].sum()

    other_runs=d15[(d15['batting_team']=='Mumbai Indians') & (d15['batsman']==i) &((d15['batsman_runs']==1) | (d15['batsman_runs']==2) | (d15['batsman_runs']==3))]['batsman_runs'].sum()

    ball_faced=d15[(d15['batting_team']=='Mumbai Indians') & (d15['batsman']==i) & (d15['extra_runs']==0)]['ball'].count()


    BIS=((tot_runs + bound_runs-(other_runs)*0.5))/ball_faced
    
    print("\n"+ i)
    print("\nTotal Batsman Runs : " + str(tot_runs))
    print("\nTotal Batsman Runs of 1s ,2s ,3s :" + str(other_runs))
    print("\nTotal Batsman Runs of Boundaries :" + str(bound_runs))
    print("\nTotal balls faced :" + str(ball_faced))
    print("\nBIS of batsman is :" + str(BIS))
    print("---------------------------------")


```
