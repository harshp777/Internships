```python
df_cpy=del_df.copy()
#df2 for matches dataset and df_cpy for deleveries datase
df2['team1'].value_counts()
```

```python
day17=df2[((df2['team1']=='Delhi Capitals') & (df2['team2']=='Sunrisers Hyderabad')) |(df2['team2']=='Delhi Capitals') & (df2['team1']=='Sunrisers Hyderabad')]
day17.head()
```

```python
#Replacing Delhi Daredevils o Delhi Capitals for deliveries dataset
df_cpy.info()
df_cpy['batting_team'].unique()

df_cpy['batting_team']=df_cpy['batting_team'].replace('Delhi Daredevils','Delhi Capitals')

df_cpy['bowling_team'].value_counts()
df_cpy['bowling_team']=df_cpy['bowling_team'].replace('Delhi Daredevils','Delhi Capitals')
```

```python
#No of matches played between them
matches=df_cpy[((df_cpy['batting_team']=='Delhi Capitals') & (df_cpy['bowling_team']=='Sunrisers Hyderabad')) |(df_cpy['batting_team']=='Sunrisers Hyderabad') & (df_cpy['bowling_team']=='Delhi Capitals')]['match_id'].nunique()
matches
```

```python
#Head to Head 

print("No of times SRH won the match : "+ str(len(day17[day17['winner']=='Sunrisers Hyderabad'])))
print("No of times DC won the match : "+ str(len(day17[day17['winner']=='Delhi Capitals'])))
```

```python
d17=df_cpy[((df_cpy['batting_team']=='Delhi Capitals') & (df_cpy['bowling_team']=='Sunrisers Hyderabad')) |((df_cpy['batting_team']=='Sunrisers Hyderabad') & (df_cpy['bowling_team']=='Delhi Capitals'))  ]
d17.head()
```

```python
#RUNS Totally sscored in a match

#Lets calculate total runs scored by both teams collectively in all matches 

dc_runs=d17[(d17['batting_team']=='Delhi Capitals')]['total_runs'].sum()
print("DC total runs : " + str(dc_runs))

srh_runs=d17[(d17['batting_team']=='Sunrisers Hyderabad')]['total_runs'].sum()
print("SRH  total runs : " + str(srh_runs))

print("Total runs of both the teams collectively : " + str(dc_runs + srh_runs))

print("\n---------------------------------------------------------------------")

#Average total of both the teams 

avg=(dc_runs + srh_runs)/matches

print("\nAverage total runs of a match is : " + str(avg))
```

```python
#total wickets of caught out in a match
dc_wik=d17[(d17['batting_team']=='Delhi Capitals') & ((d17['dismissal_kind']=='caught') | (d17['dismissal_kind']=='caught and bowled'))]['dismissal_kind'].count()
print("Total wickets of caught : " + str(dc_wik))

srh_wik=d17[(d17['batting_team']=='Sunrisers Hyderabad') & ((d17['dismissal_kind']=='caught') | (d17['dismissal_kind']=='caught and bowled'))]['dismissal_kind'].count()
print("Total wickets of caught : " + str(srh_wik))


#avg wickets 

avg=(dc_wik+srh_wik)/matches
print("Average wickets lost of caught in a match :"+ str(avg))
```

```python
#SRH RUNS 

srh_runs = d17[(d17['batting_team']=='Sunrisers Hyderabad') & (d17['over']<=6)]['total_runs'].sum()
print("Total SRH runs : " + str(srh_runs))
print("AVg DC runs :" + str(srh_runs/matches))


#DC RUNS 

dc_runs = d17[(d17['batting_team']=='Delhi Capitals') & (d17['over']<=6)]['total_runs'].sum()
print("\nTotal DC runs : " + str(dc_runs))
print("Avg DC runs:" + str(dc_runs/matches))



```

```python
#OVERALL

#SRH RUNS 
srh_match=df_cpy[df_cpy['batting_team']=='Sunrisers Hyderabad']['match_id'].nunique()
srh_runs = df_cpy[(df_cpy['batting_team']=='Sunrisers Hyderabad') & (df_cpy['over']<=6)]['total_runs'].sum()
print("Total SRH runs : " + str(srh_runs))
print("Avg SRH runs :" + str(srh_runs/srh_match))
#dc RUNS 

dc_match=df_cpy[df_cpy['batting_team']=='Delhi Capitals']['match_id'].nunique()
dc_runs = df_cpy[(df_cpy['batting_team']=='Delhi Capitals') & (df_cpy['over']<=6)]['total_runs'].sum()
print("\nTotal DC runs : " + str(dc_runs))
print("Avg DC runs:" + str(dc_runs/dc_match))


```

```python
#EXTRAS 

dc_ex=d17[(d17['bowling_team']=="Delhi Capitals") & (d17['extra_runs']!=0)]['extra_runs'].sum()
print("DC : "+ str(dc_ex/matches))

srh_ex=d17[(d17['bowling_team']=="Sunrisers Hyderabad") & (d17['extra_runs']!=0)]['extra_runs'].sum()
print("SRH : "+ str(srh_ex/matches))
```

```python
#OVERALL 

dc_ex=df_cpy[(df_cpy['bowling_team']=="Delhi Capitals") & (df_cpy['extra_runs']!=0)]['extra_runs'].sum()
print("DC : "+ str(dc_ex/dc_match))

srh_ex=df_cpy[(df_cpy['bowling_team']=="Sunrisers Hyderabad") & (df_cpy['extra_runs']!=0)]['extra_runs'].sum()
print("SRH : "+ str(srh_ex/srh_match))
```
